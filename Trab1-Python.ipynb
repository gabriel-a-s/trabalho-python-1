```python
## Questão5- Agrupar os países por ordem decrescente da soma de turistas nos meses e anos selecionados. Selecionar os ANOS onde os valores de 
#'col=="22 and over" e Key characteristic=="Length of stay"'estiverem acima da média total de cada ano e os MESES onde os valores de 
#'col=="Melbourne" e Key characteristic=="Closest Port of flight"'estiverem acima da média total de cada mês. Desconsiderar o período "Total".
```

```python
#Selecionar os anos 
```

```python
a=tabela_caract2[(tabela_caract2["Key characteristic"]=="Length of stay") & (tabela_caract2['col'] == "22 and over")& (tabela_caract2['Period'] != "Total")]
```

```python
a
```

```python
a=a.groupby(['col'])['2013','2014','2015','2016','2017','2018'].agg(['mean'])
```

```python
a1=(a.iloc[0,:][(a.iloc[0,:]>a.iloc[0,:].values.mean()).values]).reset_index(inplace=False)
```

```python
a
```

```python
a.to_html('fig7')
```

```python
#Selecionar os meses
```

```python
m=tabela_caract2[(tabela_caract2["Key characteristic"]=="Closest Port of flight") & (tabela_caract2['col'] == "Melbourne")& (tabela_caract2['Period'] != "Total")]
```

```python
m=pd.melt(m, id_vars=['Period'], value_vars=['2013','2014','2015','2016','2017','2018'],var_name='Var', value_name='Val')
```

```python
m=m.set_index(['Period']).groupby(level = ['Period']).mean().sort_values('Val', ascending=False)
```

```python
m=m.reset_index(inplace=False)
```

```python
m=pd.DataFrame(m,columns = ["Period","Val","Logic"] )
```

```python
m["Logic"]=m["Val"]>(m["Val"]).mean()
```

```python
m
```

```python
m.to_html('fig9')
```

```python
meses=list(m.loc[m['Logic'] == True,"Period"])
```
